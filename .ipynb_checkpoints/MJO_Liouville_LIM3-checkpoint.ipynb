## Develop LIM for forecasting 
$\mathbf{G}$ is the linear dynamical operator 
<br/>
$\mathbf{\epsilon}$ is the white noise forcing, which is defined as $\epsilon\epsilon^{T}=(\mathbf{x_{t+1}}-\mathbf{ax_{t}})(\mathbf{x_{t+1}}-\mathbf{ax_{t}})^{T}$ according to fluctuation-dissipation relationship

```python
total_lag       = 1
rmm             = np.concatenate((np.matrix(rmm1), np.matrix(rmm2)), axis=0)
predictor       = rmm[:,:-total_lag]
predictant      = rmm[:,total_lag:]
reg             = 1e1
xTx             = predictor.dot(predictor.T)+ reg*np.eye(total_lag*2)
G               = predictant.dot(predictor.T).dot(np.linalg.inv(xTx))      


epsilon_squared = np.zeros((401,401,2,2)) # state-dependent white noise forcing 
xi = zi         = np.arange(-2,2.01,0.01)
xi,zi           = np.meshgrid(xi,zi)

for i in np.arange(0,401,1):
    for j in np.arange(0,401,1):
        distance  = (rmm1-(i-200)/100)**2+(rmm2-(j-200)/100)**2 # range from x=[-2,2] and y=[-2,2]
        threshold = np.percentile(distance,5)
        posi      = np.array(np.where(distance<threshold))
        posi      = np.squeeze(posi.T[:-2])
        # epsilon_squared = (predictant - G.dot(predictor)).dot((predictant - G.dot(predictor)).T)/np.shape(rmm1)[0]
        epsilon_squared[j,i,:,:] = (predictant[:,posi] - G.dot(predictor[:,posi])).dot((predictant[:,posi] - G.dot(predictor[:,posi])).T)/np.size(posi)

```

```python
print(datetime.datetime.today())

xi = zi       = np.arange(-2,2.01,0.01)
xi,zi         = np.meshgrid(xi,zi)


def residual_forcing(xi,zi,epsilon_squared,location,ens):
    sigma_rmm1     = griddata((xi.flatten(),zi.flatten()), epsilon_squared[:,:,0,0].flatten(), (location[0], location[1]), method='nearest')
    sigma_rmm2     = griddata((xi.flatten(),zi.flatten()), epsilon_squared[:,:,1,1].flatten(), (location[0], location[1]), method='nearest')
    sigma_rmm1rmm2 = griddata((xi.flatten(),zi.flatten()), epsilon_squared[:,:,0,1].flatten(), (location[0], location[1]), method='nearest')
    sigma_rmm2rmm1 = griddata((xi.flatten(),zi.flatten()), epsilon_squared[:,:,1,0].flatten(), (location[0], location[1]), method='nearest')
    cov            = [[sigma_rmm1,sigma_rmm1rmm2],[sigma_rmm2rmm1,sigma_rmm2]]
    cov            = np.squeeze(cov)
    #print(np.shape(cov))
    forcing        = np.zeros((2,ens))
    for i in range(ens):
        forcing[:,i] = np.random.multivariate_normal([0,0], np.squeeze(cov[:,:]), size=1)
  
    return forcing


# # with a total of 50000 steps 
# rmm1_ML        = np.zeros((50000,))
# rmm2_ML        = np.zeros((50000,))

# for i in range(50000):
#     if i==0:
#         # random_data     = np.random.multivariate_normal([0,0],epsilon_squared, 1)
#         predictor       = G.dot(np.matrix(predictor[:,0]))+residual_forcing(xi,zi,epsilon_squared,predictor[:,0]).T
#     else:
#         # random_data     = np.random.multivariate_normal([0,0],epsilon_squared, 1)
#         predictor       = G.dot(np.matrix(predictor))+residual_forcing(xi,zi,epsilon_squared,predictor[:,0]).T
#     rmm1_ML[i] = np.array(predictor)[0]
#     rmm2_ML[i] = np.array(predictor)[1]
```

## Calculating the determinant of Jacobian matrix
## to make it state-dependent, we choose the following form

for a system of $\mathbf{x_{t+1}}=\mathbf{Gx_{t}}+\mathbf{\mathcal{N}(0,\epsilon\epsilon^T)}$

The corresponding determinant = $\mathbf{det}(\mathbf{x}_{t+1}\mathbf{x}_{t+1}^T [\mathbf{x}_{t}\mathbf{x}_{t+1}^T]^{-1})$

```python
# using 5000 ensemble for estimation 
total_lag       = 1
rmm             = np.concatenate((np.matrix(rmm1), np.matrix(rmm2)), axis=0)
predictor       = rmm[:,:-total_lag]
ens             = 100
det_LIM         = np.zeros((predictor.shape[1]))
#for i in range(predictor.shape[1]):
for i in range(predictor.shape[1]):
    #print(i)
    reg            = 1e-6
    x_t            = np.repeat(predictor[:,i], ens,axis=1) + np.random.normal(size=(2,ens))
    x_t_1          = G.dot(x_t)+residual_forcing(xi,zi,epsilon_squared,predictor[:,i],ens)   
    G_template     = x_t_1.dot(x_t_1.T).dot(np.linalg.inv(x_t.dot(x_t_1.T) + reg*np.eye(total_lag*2)))
    det_LIM[i]     = np.linalg.det(G_template)
    #print(det_LIM[i])
    
```

## Estimating the smoothed determinant using Fourier transform in polar coordinate
### $\mathrm{det(G)} = \sum_{j=1}^{n} r_j \mathrm{e}^{i\theta}$ 
### $j=$circle index, $r=$ radius of $jth$ circle, $\theta=$ polar angle, 

```python
from scipy.interpolate import griddata

# using Fourier transform instead 
r          = np.reshape(np.linspace(0, 3, 15, endpoint=True),[15,1])
degree     = np.reshape(np.linspace(0, 2*np.pi, 1000, endpoint=True),[1,1000])
x_ref      = np.matmul(r,np.cos(degree))
y_ref      = np.matmul(r,np.sin(degree))

det_polar  = griddata((xi.flatten(),zi.flatten()), (det_LIM_grid).flatten(), (x_ref.flatten(), y_ref.flatten()), method='nearest')
det_polar          = np.reshape(det_polar,[15,1000])
det_polar_anomaly  = np.zeros(np.shape(det_polar))
# det_polar_anomaly  = np.reshape(det_polar,[15,1000])



def fourier_anomaly(x, number_of_harmonics):
    time_series = np.zeros((np.size(x),number_of_harmonics))
    for i in range(0,number_of_harmonics):
        time_series[:,i] = np.arange(0,np.size(x))/np.size(x)*2*i*3.14159
    cos_coef    = np.matmul(np.transpose(np.cos(time_series)),x-np.mean(x))   
    sin_coef    = np.matmul(np.transpose(np.sin(time_series)),x-np.mean(x))
    without_n0  = (np.matmul(np.cos(time_series),cos_coef)+np.matmul(np.sin(time_series),sin_coef))/np.size(x)  
    return without_n0

for i in range(15):
    det_polar_anomaly[i,:] = fourier_anomaly(det_polar[i,:],3)
for i in range(15):
    det_polar[i,:] = fourier_anomaly(det_polar[i,:],3)
    
    

plt.figure()
cs=plt.contourf(x_ref,y_ref,det_polar_anomaly,levels=np.arange(-0.002,0.0021,0.0001),cmap='RdBu_r',extend='both')
plt.contour(x_ref,y_ref,det_polar_anomaly,levels=[0],cmap='RdBu_r',extend='both')
#plt.contour(xi,zi,epsilon_squared[:,:,0,0]+epsilon_squared[:,:,1,1],colors='k')
plt.colorbar(cs)

```

```python
# calculating 95% confidence interval 

def find_CI(rho):
    global CI_95,CI_66, posi
    for i in range(100):
        value_1     = np.percentile(rho/np.sum(rho),100-i-1)
        value_2     = np.percentile(rho/np.sum(rho),100-i)
        posi_1      = np.where((rho/np.sum(rho))>value_1)
        posi_2      = np.where((rho/np.sum(rho))>value_2)
        threshold_1 = np.sum((rho/np.sum(rho))[posi_1])
        threshold_2 = np.sum((rho/np.sum(rho))[posi_2])
        if (threshold_1-0.95)*(threshold_2-0.95)<0:
            posi    = posi_2
            indices = np.random.choice(range(len(posi_2[0])), replace=False, size=30)
            posi    = np.array(posi_2[0])[indices.astype(int)]
            CI_95   = value_2
        elif  (threshold_1-0.66)*(threshold_2-0.66)<0:
            posi    = posi_2
            indices = np.random.choice(range(len(posi_2[0])), replace=False, size=30)
            CI_66   = value_2
    return CI_95, CI_66, posi
```

```python
# make 500 times of reforecast 

iteration = 0
count_95  = np.zeros((500,8))
count_66  = np.zeros((500,8))

for iteration in range(1):
    print('working on iteration'+str(iteration))
    count = 0
    for count in range(1000):
        a     = np.random.normal(loc=0.0, scale=1, size=None)
        b     = np.random.normal(loc=0.0, scale=1, size=None)
        if np.abs(a) <1 and np.abs(b)<1:
            x_coordinate  = a
            y_coordinate  = b
            break
    
    from scipy.ndimage import gaussian_filter
    # calculate the forecast density from LIM
    # create initial rho but using machine learning data
    #x_coordinate  = a
    #y_coordinate  = b
    # target grid to interpolate to
    xi = zi       = np.arange(-2,2.01,0.01)
    xi,zi         = np.meshgrid(xi,zi)
    rmm1_ML       = (rmm1_ML-np.nanmean(rmm1_ML))/np.nanstd(rmm1_ML)
    rmm2_ML       = (rmm2_ML-np.nanmean(rmm2_ML))/np.nanstd(rmm2_ML)


    # Step 1: 
    # assign initial PDF to data point
    var          = multivariate_normal(mean=[x_coordinate,y_coordinate], cov=[[0.1,0],[0,0.1]])
    rho_zero     = np.zeros(np.shape(rmm1_ML))
    
    for i in range(np.size(rmm1_ML)):
        rho_zero[i] = var.pdf([rmm1_ML[i],rmm2_ML[i]])
    
    # Step 2:
    # creating checker 
    xi = zi       = np.arange(-2.0,2.01,0.01)
    xi,zi         = np.meshgrid(xi,zi)
    
    # meshgrid to Cartesian coordinate 
    # meshgrid back to grid_wise 
    # these two steps help remove wiggle pattern 
    
    # then integrate forward in time! 
    t_total       = 47
    rho           = np.zeros((np.size(rmm1_ML[:-t_total,]),t_total))
    rho[:,0]      = rho_zero[:-t_total,]
    
    for t in range(t_total-1):
        if t>0:
            rho_tmp         = griddata((rmm1_ML[t:-t_total+t],rmm2_ML[t:-t_total+t]),rho[:,t],(xi,zi),method='linear')
            rho_tmp         = gaussian_filter(rho_tmp, sigma=20)
            rho[:,t]        = griddata((xi.flatten(),zi.flatten()), rho_tmp.flatten(), (rmm1_ML[t:-t_total+t], rmm2_ML[t:-t_total+t]), method='nearest')
        #det       = griddata((xi.flatten(),zi.flatten()), probability_ref.flatten(), (rmm1[t:13514-t_total+t], rmm2[t:13514-t_total+t]), method='nearest')
        det_SINDy         = griddata((x_ref.flatten(),y_ref.flatten()), det_polar.flatten(), (rmm1_ML[t:-t_total+t], rmm1_ML[t:-t_total+t]), method='nearest')
        rho[:,t+1]        = rho[:,t]/(det_SINDy)
        rho[:,t+1]        = rho[:,t+1]/np.nansum(rho[:,t+1])
        

    count2 = 0 
    for t in range(0,40,5):
        # model forecast
        rho_tmp = griddata((rmm1_ML[t:-t_total+t],rmm2_ML[t:-t_total+t]),rho[:,t]/np.nansum(rho[:,t]),(xi,zi),method='linear')
        plt.figure()
        plt.contourf(xi,zi,rho_tmp)
        CI_95,CI_66,posi_95 = find_CI(rho[:,t])
        plt.contour(xi,zi,rho_tmp,levels=[CI_95])
        plt.contour(xi,zi,rho_tmp,levels=[CI_66],colors='r')
    
        # verification
        # if today the observation is independent of the forecast above 
        # we can first select some random points falling within the CI
    
        # choose 30 data points at t = 0
        if t==0:
            # print(t)
            rho_obs_grid          = griddata((xi.flatten(),zi.flatten()),(rho_tmp).flatten(),(rmm1[t:-t_total+t],rmm2[t:-t_total+t]),method='linear')
            rho_obs_Cartesian     = griddata((rmm1[t:-t_total+t], rmm2[t:-t_total+t]), rho_obs_grid,(xi,zi),method='nearest')
            time_range            = np.arange(np.size(rho_obs_grid))
            posi_2                = np.where((rho_obs_grid>CI_95) & (time_range<13514-47))
            indices               = np.random.choice(range(len(posi_2[0])), replace=False, size=30)
            posi_95_obs           = np.array(posi_2[0])[indices.astype(int)]
            # CI_95_obs,posi_95_obs = find_CI(rho_obs_grid)
            # print(posi_95_obs)
        
        #plt.scatter((rmm1[t:-t_total+t])[posi_95_obs],(rmm2[t:-t_total+t])[posi_95_obs],s=10,color='k')
        
        # calculate the data point falling outside CI at each step 
        # step 1: find the location
        data_point            = (rmm1[t:-t_total+t])[posi_95_obs],(rmm2[t:-t_total+t])[posi_95_obs]
        # step 2: find 95% CI
        CI_95_obs,CI_66_obs,posi_95_    = find_CI(rho[:,t])
        # step 3: map the forecast rho back to 30 data points and see if they are standing on rho smaller than 95% CI
        point_wise_rho        = griddata((xi.flatten(),zi.flatten()),(rho_tmp).flatten(),data_point,method='nearest')
        # count how many points fall outside CI
        # print(sum(point_wise_rho<CI_66))
        # print(sum(point_wise_rho<CI_95))
        count_66[iteration,count2] = sum(point_wise_rho<CI_66)
        count_95[iteration,count2] = sum(point_wise_rho<CI_95)

        count2 = count2+1
        
```
